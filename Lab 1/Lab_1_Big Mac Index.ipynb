```python
# =================
# Data Source: Table 1.4 (Jan 2015)
# =================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Country': [
        'United States', 'Argentina', 'Australia', 'Brazil', 'Britain', 'China', 'Egypt', 'Euro Area', 'Honk Kong', 'Indonesia', 'Japan', 'Mexico', 'Norway', 'Pakistan', 'Philippines', 'Russia' , 'Saudi Arabia', 'South Africa', 'South Korea'
    ],
    'iso_a3': [
        'USA', 'ARG', 'AUS', 'BRA', 'GBR', 'CHN', 'EGY', 'EMU', 'HKG', 'IDN', 'JPN', 'MEX', 'NOR', 'PAK', 'PHL', 'RUS', 'SAU', 'ZAF', 'KOR'

    ],
    'local_price': [
        4.79, 28.00, 5.30, 13.50, 2.89, 17.20, 16.93, 3.68, 18.80, 27939, 370, 49.00, 48.00, 300, 163, 89.00, 11.00, 25.50, 4100
    ],
    'dollar_ex': [
        1.00, 8.62, 1.23, 2.59, 0.66, 6.21, 7.36, 0.86, 7.74, 12472, 117.83, 14.63, 7.62, 100.67, 44.41, 65.44, 3.75, 11.49, 1084.66
    ]
}

# Convert to data frame
df = pd.DataFrame(data)
print(df)
```

```python
# Load full dataset from GitHub
# Use "Raw" URL

url = "https://raw.githubusercontent.com/TheEconomist/big-mac-data/refs/heads/master/source-data/big-mac-source-data-v2.csv"

full_df = pd.read_csv(url)

#Inspect the columns
print(full_df.columns)
print(full_df.head())
```

```python
# 1. Calculate Big Mac Index
df['dollar_price'] = df['local_price'] / df['dollar_ex']

# 2. Identifying US Price (Benchmark)
us_price = df.loc[df['iso_a3'] == 'USA', 'dollar_price'].values

# 3. Calculate implied PPP
df['implied_PPP'] = df['local_price'] / us_price

# 4. Calculate Valuation
# (implied - actual)/ actual

df['valuation'] = df['implied_PPP'] - df['dollar_price']/ df['dollar_price']

# 5.Convert to percentage

df['valuation_pct'] = df['valuation'] * 100

# Display results sorted by valuation

print(df[['Country', 'dollar_price', 'valuation_pct']].sort_values('valuation_pct'))
```
